```python
from kafka import KafkaConsumer
from kafka import KafkaClient
from json import loads
from hdfs import InsecureClient
import requests
import pandas as pd
from io import StringIO
from json import dump,dumps
from datetime import datetime
API_ENDPOINT = "https://api.powerbi.com/beta/164e1b0e-c8e5-41a9-9bbb-6f7ed40eef04/datasets/09610c4f-88bf-4c46-8599-e02307294b82/rows?key=6mY%2Bj2pdecE6kxvyEigWgTs%2BpYgqwemPiPePsgxg1R7C%2BIewGp9QxvkiD1mTplJ7pd9jLBBx1Rf8Lyh4P4iA2Q%3D%3D"
import json
topic_name = 'rta'

if __name__ == "__main__":
    consumer = KafkaConsumer(topic_name,
                             bootstrap_servers='localhost:9092', 
                             value_deserializer=lambda m: json.loads(m.decode('utf-8')))
    for message in consumer:
        content = message[6]
        d = json.loads(content[1:-1])
        item_id = int(d['item_id'])
        sku = str(d['sku'])
        price = int(d['price'])
        time = datetime.now().isoformat()
        dt = pd.DataFrame({'item_id':[item_id], 'sku':[sku], 'price':[price], 'time':[time]})
        data = dt.to_json(orient='records')
        req = requests.post(API_ENDPOINT, data)
        print(req)
        
```
